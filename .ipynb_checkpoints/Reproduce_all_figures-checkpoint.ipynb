```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import IFrame
from IPython.display import Image
%config InlineBackend.figure_format = 'retina'
evalNums = 30
trainIters = 300000
resetNum = 50
```

```python
### Figure 2: ease of teaching of artificial languages

def plot_ar():
    # load artifial languages
    trials = 100
    permuted_mean_l = [[] for _ in range(2)]
    permuted_ste_l = [[] for _ in range(2)] 
    for f in ['permuted']:
        trainAcc = np.zeros((trials, 1000 * 30))
        evalAcc = np.zeros((trials, 30, 1000))
        dtopo = np.zeros(trials)
        for n in range(trials):
            trainAcc[n] = np.load(f + '/' + f + '_trial' + str(n) + '/trainAcc.npy')
            for j in range(30):
                evalAcc[n][j] = trainAcc[n][j*1000: (j+1)*1000]
            dtopo[n] = np.load(f + '/' + f + '_trial' + str(n) + '/dTopo.npy')
        evalAcc_mean = np.mean(evalAcc, 1)
        measure = [evalAcc_mean, dtopo]
        for ind, m in enumerate(measure):
            measure_mean = np.mean(m, 0)
            measure_std = np.std(m, 0)
            measure_ste = measure_std / np.sqrt(trials)
            permuted_mean_l[ind].append(measure_mean)
            permuted_ste_l[ind].append(measure_ste)
    permuted_evalAcc_mean, permuted_topo_mean = permuted_mean_l
    permuted_evalAcc_ste, permuted_topo_ste = permuted_ste_l

    perfect_mean_l = []
    perfect_ste_l = [] 
    perfect_dtopo = []
    for f in ['perfect']:
        trainAcc = np.load(f + '/trainAcc.npy')
        evalAcc = np.zeros((evalNums, 1000))
        for j in range(evalNums):
            evalAcc[j] = trainAcc[j*1000 : (j+1)*1000]
        dtopo = np.load(f + '/dTopo.npy')

        evalAcc_mean = np.mean(evalAcc, 0)
        evalAcc_std = np.std(evalAcc, 0)
        evalAcc_ste = evalAcc_std / np.sqrt(30)

        perfect_mean_l.append(evalAcc_mean)
        perfect_ste_l.append(evalAcc_ste)
        perfect_dtopo.append(dtopo)

    ar_mean = perfect_mean_l + permuted_evalAcc_mean
    ar_ste = perfect_ste_l + permuted_evalAcc_ste
    ar_topo = perfect_dtopo + permuted_topo_mean

    # plot Figure 2
    fig = plt.figure()
    evalRange = np.arange(1, 1000)
    c = ['r', 'y']
    fn = ['perfect', 'permuted', ]
    for i in range(len(fn)): 
        plt.plot(evalRange, [ar_mean[i][j] for j in evalRange], c[i], label=fn[i] + ' (Topo={:.2f})'.format(ar_topo[i]))
        plt.fill_between(evalRange, [ar_mean[i][j]+ar_ste[i][j] for j in evalRange], [ar_mean[i][j]-ar_ste[i][j] for j in evalRange], facecolor=c[i], alpha=0.2)
    plt.xlabel('Number of Iterations', fontsize='14')
    plt.ylabel('Teaching Accuracy(%)', fontsize='14')
    plt.legend()
    plt.show()
    fig.savefig('ar_speed.pdf')
    plt.close()
```

```python
### Figure 3: ease of teaching of emergent languages
### Figure 4: ease of teaching of emergent languages during training
### Figure 5: topographic similarity of emergent languages
def get_seq_line_name(name):
    if name[-7:] == 'noreset':
        lname = 'no-reset'
    else:
        lname = 'reset'
    return lname
    
def seq_plot(fname, color, trials, suf='', moreColor=None, seq=range(0,5)):
    # aggregate results    
    # 'trainAcc', 'evalAcc', 'dtopo'
    measure_mean_list = [[] for _ in range(3)] # [method][measure]
    measure_ste_list = [[] for _ in range(3)]

    for f in fname:
        trainAcc = np.zeros((trials, trainIters))
        evalAcc = np.zeros((trials, 5, evalNums, 1000))
        evalAcc_mean = np.zeros((trials, 5, 1000))
        dtopo = np.zeros((trials, resetNum+1))
        prefix = ''
        suffix = suf
        for n in range(trials):
            trainAcc[n] = np.load(prefix + f + suffix + '/' + prefix + f + '_trial' + str(n) + '/trainAcc.npy')
            dtopo[n] = np.load(prefix + f + suffix + '/' + prefix + f + '_trial' + str(n) + '/dTopo.npy')
            evalAcc[n] = np.load(prefix + f + suffix + '/' + prefix + f + '_trial' + str(n) + '/evalAcc.npy')
        evalAcc_mean = np.mean(evalAcc, 2)
        measure = [trainAcc, evalAcc_mean, dtopo]
        for ind, m in enumerate(measure):
            measure_mean = np.mean(m, 0)
            measure_std = np.std(m, 0)
            measure_ste = measure_std / np.sqrt(trials)
            measure_mean_list[ind].append(measure_mean)
            measure_ste_list[ind].append(measure_ste)

    trainAcc_mean_l, evalAcc_mean_l, dTopo_mean_l, = measure_mean_list
    trainAcc_ste_l, evalAcc_ste_l, dTopo_ste_l, = measure_ste_list
    
    # plot ease-of teaching 
    fig = plt.figure()
    evalRange = np.arange(1, 1000)
    c = ['r', 'y']
    fn = ['perfect', 'permuted', ]
    for i in [0]: 
        plt.plot(evalRange, [ar_mean[i][j] for j in evalRange], c[i], label=fn[i] + ' (Topo={:.2f})'.format(ar_topo[i]))
        plt.fill_between(evalRange, [ar_mean[i][j]+ar_ste[i][j] for j in evalRange], [ar_mean[i][j]-ar_ste[i][j] for j in evalRange], facecolor=c[i], alpha=0.2)
    for i in range(len(fname)): 
        for k in [4]:
            plt.plot(evalRange, [evalAcc_mean_l[i][k][j] for j in evalRange], color[i], label=get_seq_line_name(fname[i]) + ' (Topo={:.2f})'.format(dTopo_mean_l[i][(k+1)*10]))
            plt.fill_between(evalRange, [evalAcc_mean_l[i][k][j]+evalAcc_ste_l[i][k][j] for j in evalRange], [evalAcc_mean_l[i][k][j]-evalAcc_ste_l[i][k][j] for j in evalRange], facecolor=color[i], alpha=0.2)
    for i in [1]: 
        plt.plot(evalRange, [ar_mean[i][j] for j in evalRange], c[i], label=fn[i] + ' (Topo={:.2f})'.format(ar_topo[i]))
        plt.fill_between(evalRange, [ar_mean[i][j]+ar_ste[i][j] for j in evalRange], [ar_mean[i][j]-ar_ste[i][j] for j in evalRange], facecolor=c[i], alpha=0.2)
    plt.xlabel('Number of Iterations', fontsize='14')
    plt.ylabel('Teaching Accuracy(%)', fontsize='14')
    plt.legend()
    plt.show()
    fig.savefig('seq_teachAcc_legend.pdf')
    plt.close()

    if moreColor:
        fig = plt.figure()
        evalRange = np.arange(0, 1000)
        for i in range(len(fname)):
            if i == 0:
                seq_l = seq[::-1]
            else:
                seq_l = seq
            for ind, k in enumerate(seq_l):
                plt.plot(evalRange, evalAcc_mean_l[i][k], moreColor[i][ind], label=get_seq_line_name(fname[i]) + ' ' + str((k+1)*60) + 'k' + ' (Topo={:.2f})'.format(dTopo_mean_l[i][(k+1)*10]))
                plt.fill_between(evalRange, [evalAcc_mean_l[i][k][j]+evalAcc_ste_l[i][k][j] for j in evalRange], [evalAcc_mean_l[i][k][j]-evalAcc_ste_l[i][k][j] for j in evalRange], facecolor=color[i], alpha=0.2)
        plt.xlabel('Number of Iterations', fontsize='14')
        plt.ylabel('Teaching Accuracy(%)', fontsize='14')
        plt.legend()
        plt.show()
        fig.savefig('seq_speed.pdf')
        plt.close()

    trainRange = np.arange(6000, trainIters+1, resetIters)
    partRange = np.arange(1, resetNum+1)
    fig = plt.figure()
    fig.patch.set_facecolor('xkcd:white')
    for i in range(len(fname)):
        plt.plot(trainRange, [dTopo_mean_l[i][j] for j in partRange], color[i], label=get_seq_line_name(fname[i]))
        plt.fill_between(trainRange, [dTopo_mean_l[i][j]-dTopo_ste_l[i][j] for j in partRange], [dTopo_mean_l[i][j]+dTopo_ste_l[i][j] for j in partRange], facecolor=color[i], alpha=0.2)
    plt.xlabel('Number of Iterations', fontsize='14')
    plt.ylabel('Topographic Similarity', fontsize='14')
    plt.legend()
    plt.show()
    fig.savefig('seq_topo.pdf')
    plt.close()
```

```python
### Figure 6: ease of teaching of emergent languages under population reset-one
### Figure 7: topographic similarity under population reset-one
### Figure 12: training accuracy under population reset-one
### Figure 13: entropy under population reset-one

### Figure 8: ease of teaching in population of 2 listeners
### Figure 9: topographic similarity in population of 2 listeners

### Figure 10: ease of teaching in population of 10 listeners
### Figure 11: topographic similarity in population of 11 listeners

def get_line_name(name):
    if name[:3] == 'seq':
        if name[-7:] == 'noreset':
            lname = 'no-reset'
        elif name[-5:] == 'reset':
            lname = 'reset'
    else:
        if 'noreset' in name:
            lname = 'no-reset'
        elif 'resetall' in name:
            lname = 'reset all'
        else:
            lname = 'reset 1'
        if 'l2' in name:
            lname += ' of 2 listeners'
        elif 'l10' in name:
            lname += ' of 10 listeners'
    return lname
    
def plot(fname, color, trials, suf='', analysis=False):
    # aggregate results    
    # 'trainAcc', 'evalAcc', 'dtopo', 'entropy',
    measure_mean_list = [[] for _ in range(4)] # [method][measure]
    measure_ste_list = [[] for _ in range(4)]

    for f in fname:
        trainAcc = np.zeros((trials, trainIters))
        evalAcc = np.zeros((trials, 5, evalNums, 1000))
        evalAcc_mean = np.zeros((trials, 5, 1000))
        dtopo = np.zeros((trials, resetNum+1))
        entropy = np.zeros((trials, trainIters))
        prefix = ''
        suffix = suf
        for n in range(trials):
            if f[:3] == 'pop':
                trainAcc[n] = np.load(prefix + f + suffix + '/' + prefix + f + '_trial' + str(n) + '/pop_trainAcc.npy')
            else:
                trainAcc[n] = np.load(prefix + f + suffix + '/' + prefix + f + '_trial' + str(n) + '/trainAcc.npy')
            dtopo[n] = np.load(prefix + f + suffix + '/' + prefix + f + '_trial' + str(n) + '/dTopo.npy')
            entropy[n] = np.load(prefix + f + suffix + '/' + prefix + f + '_trial' + str(n) + '/entropy.npy')
            evalAcc[n] = np.load(prefix + f + suffix + '/' + prefix + f + '_trial' + str(n) + '/evalAcc.npy')
        evalAcc_mean = np.mean(evalAcc, 2)
        measure = [trainAcc, evalAcc_mean, dtopo, entropy]
        for ind, m in enumerate(measure):
            measure_mean = np.mean(m, 0)
            measure_std = np.std(m, 0)
            measure_ste = measure_std / np.sqrt(trials)
            measure_mean_list[ind].append(measure_mean)
            measure_ste_list[ind].append(measure_ste)

    trainAcc_mean_l, evalAcc_mean_l, dTopo_mean_l, entropy_mean_l = measure_mean_list
    trainAcc_ste_l, evalAcc_ste_l, dTopo_ste_l, entropy_ste_l = measure_ste_list
    
    fig = plt.figure()
    evalRange = np.arange(0, 1000)
    for i in range(len(fname)): 
        for k in [4]:
            plt.plot(evalRange, evalAcc_mean_l[i][k], color[i], label=get_line_name(fname[i]) + ' (Topo={:.2f})'.format(dTopo_mean_l[i][(k+1)*10]))
            plt.fill_between(evalRange, [evalAcc_mean_l[i][k][j]+evalAcc_ste_l[i][k][j] for j in evalRange], [evalAcc_mean_l[i][k][j]-evalAcc_ste_l[i][k][j] for j in evalRange], facecolor=color[i], alpha=0.2)
    plt.xlabel('Number of Iterations', fontsize='14')
    plt.ylabel('Teaching Accuracy(%)', fontsize='14')
    plt.legend()
    plt.show()
    fig.savefig('teachAcc.pdf')
    plt.close()
    
    trainRange = np.arange(6000, trainIters+1, resetIters)
    partRange = np.arange(1, resetNum+1)
    fig = plt.figure()
    fig.patch.set_facecolor('xkcd:white')
    for i in range(len(fname)):
        plt.plot(trainRange, [dTopo_mean_l[i][j] for j in partRange], color[i], label=get_line_name(fname[i]))
        plt.fill_between(trainRange, [dTopo_mean_l[i][j]-dTopo_ste_l[i][j] for j in partRange], [dTopo_mean_l[i][j]+dTopo_ste_l[i][j] for j in partRange], facecolor=color[i], alpha=0.2)
    plt.xlabel('Number of Iterations', fontsize='14')
    plt.ylabel('Topographic Similarity', fontsize='14')
    plt.legend()
    plt.show()
    fig.savefig('topo.pdf')
    plt.close()
    
    if analysis:
        fig = plt.figure()
        trainRange = np.arange(0, trainIters)
        for i in range(len(fname)): 
            plt.plot(trainRange, trainAcc_mean_l[i], color[i], label=get_line_name(fname[i]))
            plt.fill_between(trainRange, [trainAcc_mean_l[i][j]+trainAcc_ste_l[i][j] for j in trainRange], [trainAcc_mean_l[i][j]-trainAcc_ste_l[i][j] for j in trainRange], facecolor=color[i], alpha=0.2)
        plt.xlabel('Number of Iterations', fontsize='14')
        plt.ylabel('Training Accuracy(%)', fontsize='14')
        plt.legend()
        plt.show()
        fig.savefig('trainAcc.png', dpi=300)
        plt.close()

        trainRange = np.arange(0, trainIters)
        fig = plt.figure()
        fig.patch.set_facecolor('xkcd:white')
        for i in range(len(fname)):
            plt.plot(trainRange, [entropy_mean_l[i][j] for j in trainRange], color[i], label=get_line_name(fname[i]))
            plt.fill_between(trainRange, [entropy_mean_l[i][j]-entropy_ste_l[i][j] for j in trainRange], [entropy_mean_l[i][j]+entropy_ste_l[i][j] for j in trainRange], facecolor=color[i], alpha=0.2)
        plt.xlabel('Number of Iterations', fontsize='14')
        plt.ylabel('Entropy', fontsize='14')
        plt.legend()
        plt.show()
        fig.savefig('entropy.png', dpi=300)
        plt.close()
```
